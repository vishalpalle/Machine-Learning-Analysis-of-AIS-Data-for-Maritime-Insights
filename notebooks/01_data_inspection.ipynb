```python
import pandas as pd

# 1. Convert 'BaseDateTime' → datetime, specifying the format
df_raw['BaseDateTime'] = pd.to_datetime(
    df_raw['BaseDateTime'],
    format='%Y-%m-%dT%H:%M:%S',
    errors='coerce'
)

# 2. Check if any rows failed to parse
print("Parsing failures:", df_raw['BaseDateTime'].isna().sum())

# 3. Confirm dtype change and view a few entries
print(df_raw['BaseDateTime'].dtype)
df_raw[['BaseDateTime']].head(5)
```
